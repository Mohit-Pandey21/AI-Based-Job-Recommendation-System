```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("Combined_Jobs_Final.csv")

print("Dataset shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
industry_counts = df["Industry"].value_counts()
industry_percent = (industry_counts / len(df)) * 100

industry_bias_df = pd.DataFrame({
    "Industry": industry_counts.index,
    "Count": industry_counts.values,
    "Percentage": industry_percent.values
})

print(industry_bias_df.head(10))
```

```python
plt.figure(figsize=(10, 5))
industry_counts.head(10).plot(kind="bar")
plt.title("Top 10 Industries by Job Count")
plt.xlabel("Industry")
plt.ylabel("Number of Jobs")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
city_counts = df["City"].value_counts()
city_percent = (city_counts / len(df)) * 100

city_bias_df = pd.DataFrame({
    "City": city_counts.index,
    "Count": city_counts.values,
    "Percentage": city_percent.values
})

print(city_bias_df.head(10))
```

```python
plt.figure(figsize=(10, 5))
city_counts.head(10).plot(kind="bar", color="orange")
plt.title("Top 10 Cities by Job Count")
plt.xlabel("City")
plt.ylabel("Number of Jobs")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
position_counts = df["Position"].value_counts()
position_percent = (position_counts / len(df)) * 100

position_bias_df = pd.DataFrame({
    "Job Position": position_counts.index,
    "Count": position_counts.values,
    "Percentage": position_percent.values
})

print(position_bias_df.head(10))
```

```python
plt.figure(figsize=(10, 5))
position_counts.head(10).plot(kind="bar", color="green")
plt.title("Top 10 Job Positions by Frequency")
plt.xlabel("Job Position")
plt.ylabel("Number of Jobs")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
industry_bias_df.to_csv("industry_bias_analysis.csv", index=False)
city_bias_df.to_csv("geographical_bias_analysis.csv", index=False)
position_bias_df.to_csv("role_bias_analysis.csv", index=False)

print("Bias analysis CSV files saved successfully.")
```

```python
# Percentage covered by top 10 industries
top10_industry_coverage = industry_percent.head(10).sum()
print(f"Top 10 industries cover {top10_industry_coverage:.2f}% of all jobs")

# Percentage covered by top 10 cities
top10_city_coverage = city_percent.head(10).sum()
print(f"Top 10 cities cover {top10_city_coverage:.2f}% of all jobs")
```
